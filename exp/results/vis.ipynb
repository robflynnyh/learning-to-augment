```python
import matplotlib.pyplot as plt
def load_text(file_name):
    with open(file_name) as f:
        return f.read().splitlines()

def lmap(func, seq):
    return list(map(func, seq))

def extract_data(lines):
    data = []
    for line in lines:
        fields = line.split('-')
        cur_data = {}
        for field in fields:
            key, value = lmap(str.strip, field.split(':'))
            cur_data[key] = value
        data.append(cur_data)
    return data
```

Oracle Evals

```python
oracle_eval_path = './RMM/oracle/tedlium.txt'
oracle_eval_data = load_text(oracle_eval_path)
oracle_eval_data = extract_data(oracle_eval_data)
original_wer = float(oracle_eval_data[0]['Original_WER']) * 100

repeats = [int(el['Repeats']) for el in oracle_eval_data]
wers = [float(el['Updated_WER'])*100 for el in oracle_eval_data]
# marker size 1.2
plt.plot(repeats, wers, label='Oracle WER', marker='o', markersize=2.5) 
# add grid with opacity 0.5
ax = plt.gca()
ax.grid(True, which='both', linewidth=0.5, color='gray', alpha=0.3)

plt.xlabel('Augmentation Search Breadth Per Step')
plt.ylabel('WER')
# add original wer as text
plt.text(5, 8.76, f'Original WER: {original_wer:.2f}', fontsize=8)
plt.savefig('figures/oracle_wer.pdf')
```
